# 1.Import Necessary Libraries| 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
import warnings
warnings.filterwarnings('ignore')
```

# 2. Import Data

```python
My_Movies=pd.read_csv('my_movies.csv')
My_Movies
```

# 3. Data understanding 

```python
My_Movies.shape
```

```python
My_Movies.info()
```

```python
My_Movies2=My_Movies.iloc[:,5:]
My_Movies2
```

# 4.Model Building

##  4.1 Association Rule with 10% Support & 70% Confidence

```python
# With 10% Support
Frequent_Itemset=apriori(My_Movies2,min_support=0.1,use_colnames=True)
Frequent_Itemset
```

```python
# Now,with 70% Confidence
Rules=association_rules(Frequent_Itemset,metric='lift',min_threshold=0.7)
Rules
```

```python
#Identifting lift ratio > 1 is a good influential rule in selecting association rule
Rules[Rules.lift>1]
```

# Visualizing the Obtained Rule 

```python
plt.scatter(Rules['support'],Rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('With 10% Support & 70% Confidence')
plt.show()
```

##  4.2 Association Rule with 5% Support & 90% Confidence

```python
# With 5% Support
Frequent_Itemset2=apriori(My_Movies2,min_support=0.05,use_colnames=True)
Frequent_Itemset2
```

```python
# With 90% Confidence
Rules2=association_rules(Frequent_Itemset2,metric='lift',min_threshold=0.9)
Rules2
```

```python
# Finding Lift Ratio> 1 which is Good Association Rule
Rules2[Rules2.lift>1]
```

## Visualizing the Obtained rule

```python
plt.scatter(Rules2['support'],Rules2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('With 5% Support & 90% Confidence')
plt.show()
```
